```python
import yfinance as yf
import plotly.graph_objects as go
import numpy as np
```

```python
ativo = yf.download('^BVSP', period='max')['Close']
```

```python
ativo
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=ativo.index, y=ativo))

fig.update_layout(title_text='Ibovespa')


fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=ativo.index, y=ativo,
                         name='Ativo', line=dict(color='blue', width=3)))


#FHC
fig.add_vrect(x0 ='1998-01', x1='1998-12', fillcolor='#f5dd9d', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['1998-06'],y=[40000], mode='text',text='FHC', textposition='top center', showlegend=False))

#Lula
fig.add_vrect(x0='2002-01', x1='2002-12', fillcolor='#bcc499', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['2002-06'],y=[50000], mode='text',text='Lula', textposition='top center', showlegend=False))

#Lula
fig.add_vrect(x0 ='2006-01', x1='2006-12', fillcolor='#92a68a', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['2006-06'],y=[80000], mode='text',text='Lula', textposition='top center', showlegend=False))


#Dilma
fig.add_vrect(x0='2010-01', x1='2010-12', fillcolor='#7b8f8a', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['2010-06'],y=[80000], mode='text',text='Dilma', textposition='top center', showlegend=False))

#Dilma/Temer
fig.add_vrect(x0='2014-01', x1='2014-12', fillcolor='#506266', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['2014-06'],y=[80000], mode='text',text='Dilma/', textposition='top center', showlegend=False))
fig.add_trace(go.Scatter(x=['2014-06'],y=[70000], mode='text',text='Temer', textposition='top center', showlegend=False))


#Bolsonaro
fig.add_vrect(x0='2018-01', x1='2018-12', fillcolor='#d4838f', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['2018-06'],y=[100000], mode='text',text='Bolsonaro', textposition='top center', showlegend=False))


#Lula
fig.add_vrect(x0='2022-01', x1='2022-12', fillcolor='#a68c69', opacity=0.25, line_width=0)
fig.add_trace(go.Scatter(x=['2022-06'],y=[122000], mode='text',text='Lula', textposition='top center', showlegend=False))

#Layout
fig.update_layout(title='', xaxis_title='<b>Data evento', yaxis_title="<b>Ibovespa", legend_title='',
                  font=dict(family='Arial, Arial, Arial', size=20, color='black'), paper_bgcolor='#f7f8fa')

fig.update_layout(xaxis_rangeslider_visible=False, margin=dict(l=120, r=20, t=20,b=20), template='simple_white',
                  width=1200, height=500, legend=dict(orientation='h', yanchor='bottom', y=0.9, xanchor='right', x=0.15),
                  xaxis=dict(range=['1995', '2023']))
fig.show()
```
